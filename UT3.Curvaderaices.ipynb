```python
from sympy import *
import numpy as np
from control.matlab import *
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import control as ct
from IPython.display import display, Latex
```

```python
def update_plot(k):

    #sys=ct.tf([1], [1, 1])
    #sys=ct.tf([1], [1, 2, 1])
    #sys=ct.tf([1], [1, 3, 2, 0])
    #sys=ct.tf([1, 2], [1, 2, 1])
    #sys = ct.tf([1, 1], [1, 10, 100])
    sys = ct.tf([1, 3], [1, 5, 20, 16, 0])

    syscl=ct.feedback(sys*k,1)
    roots=ct.pzmap(syscl, plot=True)[0]
    #close the loop of sys

    # Plot the root locus
    ct.root_locus(sys,xlim=[-5,1],ylim=[-1,5])
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.title('Root Locus')
    plt.text(1.1, 1.2, sys, horizontalalignment='left', verticalalignment='top', transform=plt.gca().transAxes, fontsize=20)
    plt.text(1.1, 0.5, syscl, horizontalalignment='left', verticalalignment='top', transform=plt.gca().transAxes, fontsize=20)

    plt.plot(np.real(roots), np.imag(roots), 'ro')

    plt.figure(2)
    y, T = step(syscl)
    plt.plot(T, y)
    plt.grid('on')

    info = stepinfo(syscl)
    # Display step info on the step response plot
    info_text = f"Rise Time: {info['RiseTime']:.2f} s\n" \
                f"Settling Time: {info['SettlingTime']:.2f} s\n" \
                f"Overshoot: {info['Overshoot']:.2f} %\n" \
                f"Peak: {info['Peak']:.2f}\n" \
                f"Peak Time: {info['PeakTime']:.2f} s"

    plt.text(1.05, 0.95, info_text, horizontalalignment='left', verticalalignment='top', transform=plt.gca().transAxes, bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5))

    plt.show(block=True)


interact(update_plot, k=(1, 40, 0.1))
```

```python
# prompt: poner el grafico temporal al lado del grafico de curva de raices

import matplotlib.pyplot as plt
import numpy as np
plt.figure(1, figsize=(24, 24))

def update_plot(k):
    plt.figure(1)
    plt.clf()

    #sys=ct.tf([1], [1, 1])
    #sys=ct.tf([1], [1, 2, 1])
    #sys=ct.tf([1], [1, 3, 2, 0])
    #sys=ct.tf([1, 2], [1, 2, 1])
    #sys = ct.tf([1,1], [1, 10, 100])
    sys = ct.tf([1, 3], [1, 5, 20, 16, 0])


    syscl=ct.feedback(sys*k,1)
    roots=ct.pzmap(syscl, plot=False)[0] # set plot=False here

    plt.subplot(1, 2, 1)  # Left subplot for root locus
    ct.root_locus(sys,xlim=[-5,1],ylim=[-5,5], plot=True) # Plot the root locus
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.title('Root Locus')

    plt.plot(np.real(roots), np.imag(roots), 'ro')
    plt.grid(True)


    plt.subplot(1, 2, 2)  # Right subplot for step response
    y, T = step(syscl)
    plt.plot(T, y)
    plt.grid('on')
    plt.title('Step Response')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')


    info = stepinfo(syscl)
    # Display step info on the step response plot
    info_text = f"Rise Time: {info['RiseTime']:.2f} s\n" \
                f"Settling Time: {info['SettlingTime']:.2f} s\n" \
                f"Overshoot: {info['Overshoot']:.2f} %\n" \
                f"Peak: {info['Peak']:.2f}\n" \
                f"Peak Time: {info['PeakTime']:.2f} s"

    plt.text(1.1, 1, sys, horizontalalignment='left', verticalalignment='top', transform=plt.gca().transAxes, fontsize=15, bbox=dict(boxstyle='round,pad=0.3', fc='wheat', alpha=0.5))
    plt.text(1.1, 0.6, syscl, horizontalalignment='left', verticalalignment='top', transform=plt.gca().transAxes, fontsize=15, bbox=dict(boxstyle='round,pad=0.3', fc='wheat', alpha=0.5))
    plt.text(1.1, 0.2, info_text, horizontalalignment='left', verticalalignment='top', transform=plt.gca().transAxes, fontsize=15, bbox=dict(boxstyle='round,pad=0.5', fc='wheat', alpha=0.5))


    #plt.tight_layout()
    plt.show()
    #plt.rcParams['figure.figsize'] = [15, 8]

interact(update_plot, k=(1, 40, 0.1))
```
